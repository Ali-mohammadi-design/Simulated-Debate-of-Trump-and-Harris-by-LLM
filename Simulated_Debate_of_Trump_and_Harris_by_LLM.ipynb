```python
import os
openAI_API_Key=os.environ['openAI_API_Key']='Your Open AI API'
llm_config = {"model": "gpt-3.5-turbo"}
```

```python
import autogen
import pprint
```

```python
from autogen import ConversableAgent
```

```python
Donald_Trump= ConversableAgent(
    name="Donald Trump",
    system_message=
    '''Please assume the role of Donald Trump in a debate with Kamala Harris, the Democratic Party's presidential candidate.
    Respond to her arguments as Trump would, reflecting his typical speaking style and approach.''' ,
    llm_config=llm_config,
    human_input_mode="NEVER",
)

kamala_harris= ConversableAgent(
    name="kamala_harris",
    system_message=
    '''Please assume the role of kamala harris in a debate with Donald Trump, the Republican Party's presidential candidate.
    Respond to his arguments as kamala harris, reflecting her typical speaking style and approach.''',
    llm_config=llm_config,
    human_input_mode="NEVER",
)
```

```python
def debate(topic, number):
    chat_result = kamala_harris.initiate_chat(
    recipient=Donald_Trump, 
    message= f"Hey, Trump, what do you think about {topic}?",
    max_turns=number,
    summary_method="reflection_with_llm",
    summary_prompt="Please provide separate lists of key points for Trump and Harris.",
    )
    pprint.pprint(chat_result.summary)

    
```

```python
debate('Migration', 4)
```
